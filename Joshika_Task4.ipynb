**Import libraries**

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

**Load dataset**

```python
data=pd.read_csv("/content/drive/MyDrive/Oasis Intern/spam.csv",encoding='latin1')
```

```python
data_frame=pd.DataFrame(data.iloc[:,:2])
data_frame.columns=["Label","Message"]
```

```python
data_frame.head(10)
```

**Assign values for x and y**

```python
x=data_frame.iloc[:,1].values
y = data_frame.iloc[:, 0].map({'ham': 0, 'spam': 1}).values
```

```python
x
```

```python
y
```

**Split data for training and testing**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**Vectorization**

```python
vectorizer=CountVectorizer()
x_train=vectorizer.fit_transform(x_train)
x_test=vectorizer.transform(x_test)
```

**Import model**

```python
model=MultinomialNB()
```

**Train model**

```python
model.fit(x_train,y_train)
```

**Test model**

```python
y_pred=model.predict(x_test)
print(y_pred)
```

**Predict new message**

```python
msg=["Congratulations! You won Boat airpods"]
msg=vectorizer.transform(msg)
prediction=model.predict(msg)
labels = {0: "Ham", 1: "Spam"}
print(labels[prediction[0]])
```

**Calculate Accuracy**

```python
print("Accuracy:",accuracy_score(y_pred,y_test))
```
