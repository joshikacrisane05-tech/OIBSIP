**Import libraries**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
import matplotlib.pyplot as plt
```

**Load dataset**

```python
data=pd.read_csv("/content/drive/MyDrive/Oasis Intern/Advertising.csv")
```

```python
data_frame=pd.DataFrame(data.iloc[:,1:5])
data_frame
```

**Check missing values**

```python
print(data_frame.isnull().sum())
```

**Assign datas for x and y**

```python
x=data_frame.iloc[:,0:3].values
y=data_frame.iloc[:,3].values
```

```python
x
```

```python
y
```

**Split data for training and testing**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

**Model Initialization**

```python
model=LinearRegression()
```

**Model training**

```python
model.fit(x_train,y_train)
```

**Model testing**

```python
y_pred=model.predict(x_test)
print("Predicted Sales:",y_pred)
```

**Calculate errors**

```python
print("MAE:",mean_absolute_error(y_pred,y_test))
print("MSE:",mean_squared_error(y_pred,y_test))
```

**Visualization**

```python
plt.figure(figsize=(12,8))
plt.plot(y_test,label="Actual Sales",marker='o',color="orange")
plt.plot(y_pred,label="Predicted Sales",marker='*',color="purple")
plt.legend()
plt.xlabel("Test Samples")
plt.ylabel("Sales")
plt.title("Actual Sales Data Vs Predicted Sales Data")
plt.grid(True)
plt.show()
```
